The MainWindow class is a subclass of Tkinter’s tk.Tk class and thus inherits its attributes and methods, while allowing us to customize the new window objects to our needs. These new window objects will represent “notebooks” or collections of notes, and allow us to work with those notes. We are using it to represent notebooks in this application, but you might use it for other purposes in onther applications.)


The TopWindow class creates a new top window in Tkinter, which is essentially a form for entering the title, text, links, and tags for our note. When we submit the note, this “form” object has a method that creates the note’s metadata and a new Note object. That note object is appended to the list of all notes, which is an attribute of the notebook (MainWindow) class.
The Note class creates note objects that contain the  title, text, links, tags, and metadata for each note.

```python
# imports
import tkinter as tk
from tkinter import ttk
import datetime # one module for working with dates and times
import csv

created = datetime.datetime.now() # gets the current date and time
local_now = created.astimezone() # shows the local time and the GMT offset
local_tz = local_now.tzinfo 

# The MainWindow class creates a custom GUI window based on the tkinter window (tk.Tk)
# It has an __init__() method, and three additional methods (new_note(), open_notebook(), and save_notebook())
# These methods correspond to new, open, and save buttons in the window.
# The new_note method calls the NoteForm class to create a new note form top level window.

class MainWindow(tk.Tk):
    def __init__(self):  #initialize the main window
        super().__init__() # initialize it as a tkinter window
        
        # self.geometry("600x400") # set the default window size
        self.title('Notebook') #set the default window title
        self.notebook = [] # initialize an attribute named 'notebook' as an empty list
        # I think once I use the save_notebook() method, it will take all the notes from the notes list and put the notes
        # full of notes objects into one one .txt, .csv, .json, and put that one file into notebook.
        self.notes = []        
        self.column_num = 0
        self.new_notes_csv = []
        
        title_label = tk.Label(self, bg='light gray', text='Note Title:')
        title_label.grid(padx=10, pady=10, row=1, column=0, sticky='e')

        # create our note title entry field
        self.note_title = tk.Entry(self, width=30)
        self.note_title.grid(padx=10, pady=10, row=1, column=1, sticky='w')
        self.note_title.insert(0, 'New note title') # adds default text (useful during development)
        
        b1 = tk.Button(self, text='new note', command=self.new_note)
        b1.grid(padx=10, pady=10, row=5, column=1, sticky='w')

        b3 = tk.Button(self, text='save notebook', command=self.save_notebook)
        b3.grid(padx=10, pady=10, row=5, column=0)

        b4 = tk.Button(self, text='quit', command=self.destroy)
        b4.grid(padx=10, pady=10, row=6, column=0)
        
        notes_section = tk.Label(self, bg='light gray', text='Notebooks:')
        notes_section.grid(padx=10, pady=10, row=1, column=2, sticky='w')

        # add additionallines to the __init__() function
        
    # display notebooks that are already made
    def get_title(self): 
        with open('filenames.txt', 'r') as display:
            files = display.readlines()
            self.corrected = [x.rstrip('\n') for x in files]
        for file in self.corrected:
            self.column_num += 1
            self.display_notebook = tk.Button(self, text=file, command=lambda fname=file: self.open_notebook(fname))
            self.display_notebook.grid(padx=10, pady=10, row=7, column=self.column_num)
        return None
            
    def new_note(self):
        note_window = NoteForm(self, self.notebook, self.notes)
        return None

    # WRITE ALL THE INFORMATION TO ONE BUTTON, THEN PLACE IT?
    def open_notebook(self, filename):
        open_row = 0
        open_col = 0
        open_note = tk.Toplevel()
        open_note.title("Notebook")
        with open(f"{filename}.csv", "r", newline='') as csvfile:
            reader = csv.DictReader(csvfile)
            for row in reader:
                self.new_notes_csv.append(row)
        for note in self.new_notes_csv:       
            # note IS A DICTIONARY OBJ        
            title = tk.Label(open_note, bg='light gray', text=f"TITLE: {note['title']}")
            title.grid(padx=10, pady=10, row=open_row, column=open_col, sticky='w')
            open_row += 1
            title = tk.Label(open_note, text=note['text'], wraplength=200, justify='left')
            title.grid(padx=10, pady=10, row=open_row, column=open_col, sticky='w')
            open_row += 1
            title = tk.Label(open_note, text=f"LINK: {note['link']}")
            title.grid(padx=10, pady=10, row=open_row, column=open_col, sticky='w')
            open_row += 1
            title = tk.Label(open_note, text=f"TAGS: {note['tags']}")
            title.grid(padx=10, pady=10, row=open_row, column=open_col, sticky='w')
            open_row += 1
            title = tk.Label(open_note, text=note['meta'])
            title.grid(padx=10, pady=10, row=open_row, column=open_col, sticky='w')
            open_col += 1
            open_row = 0
        self.new_notes_csv = []
        return None

    def save_notebook(self):
        filename = self.note_title.get()  # Get the filename from the entry widget
        with open('filenames.txt', 'a') as file:
            file.write(f'{filename}\n')

        with open(f"{filename}.csv", "w", newline='') as csvfile:
            fieldnames = ['title', 'text', 'link', 'tags', 'meta']
            writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
            writer.writeheader()
            for note in self.notes:
                writer.writerow(note.as_dict())
        self.column_num += 1
        display_notebook = tk.Button(self, text=filename, command=lambda fname=filename: self.open_notebook(fname))
        display_notebook.grid(padx=10, pady=10, row=7, column=self.column_num)
        self.notes = []

        return None


# the NoteForm() class creates a Toplevel window that is a note form containing fields for
# data entry for title, text, link, and tags. It also calculates a meta field with date, time, and timezone
# the Noteform class has an __init__() method, and a submit() method that is called by a submit button
# the class may contain additional methods to perform tasks like calculating the metadata, for example
# the submit method calls the MakeNote class that transforms the the entered data into a new note object.

class NoteForm(tk.Toplevel):
    
    def __init__(self, master, notebook, notes): # initialize the new object
        super().__init__(master) # initialize it as a toplevel window
        self.notebook = notebook
        self.notes = notes
        self.note_dict = {}
        self.meta = f'CREATED: {created}, {local_tz}'
        
        self.geometry("600x400") # set the default window size
        self.title('New Note') #set the default window title
        
        title_label = tk.Label(self, bg='light gray', text='Note Title:')
        title_label.grid(padx=10, pady=10, row=1, column=0, sticky='e')

        text_label = tk.Label(self, bg='light gray', text='Note Text:')
        text_label.grid(padx=10, pady=10, row=2, column=0, sticky='e')

        link_label = tk.Label(self, bg='light gray', text='Link:')
        link_label.grid(padx=10, pady=10, row=3, column=0, sticky='e')

        tag_label = tk.Label(self, bg='light gray', text='Tags:')
        tag_label.grid(padx=10, pady=10, row=4, column=0, sticky='e')

        # create our note title entry field
        self.note_title = tk.Entry(self, width=80)
        self.note_title.grid(padx=10, pady=10, row=1, column=1, sticky='w')
        self.note_title.insert(0, 'New note title') # adds default text (useful during development)

        # create our note text field
        self.note_text = tk.Text(self, height=10, width=60)
        self.note_text.grid(padx=10, pady=10, row=2, column=1)
        self.note_text.insert('1.0', 'DEFAULT') # adds default text (useful during development)

        # create our note link entry field
        self.note_link = tk.Entry(self, width=80)
        self.note_link.grid(padx=10, pady=10, row=3, column=1, sticky='w')
        self.note_link.insert(0, 'If there is a link with this note enter it here. Else, enter "None"') # adds default text (useful during development)

        # create our note link entry field
        self.note_tags = tk.Entry(self, width=80)
        self.note_tags.grid(padx=10, pady=10, row=4, column=1, sticky='w')
        self.note_tags.insert(0, 'Enter #tags, separated by commas') # adds default text (useful during development)
        
        # submit button
        submit_button = tk.Button(self, text='submit', command=self.submit)
        submit_button.grid(padx=10, pady=10, row=5, column=0)

        # add additional lines to set the new window's attributes and default parameters

    def submit(self):
        # add lines to the submit method
        self.note_dict = {
        'title': self.note_title.get(),
        'text': self.note_text.get('1.0', 'end'),
        'link': self.note_link.get(),
        'tags': self.note_tags.get(),
        'meta': self.meta
        }
        
        new_note = MakeNote(self.note_dict)
        self.notes.append(new_note)
        # print(type(self.notes[0]))
        return self.note_dict

    
# The MakeNote class takes a dictionary containing the data entered into the form window (from NoteForm),
# and transforms it into a new note object.
# At present the note objects have attributes but no methods.
    
class MakeNote():
    
    def __init__(self, note_dict):
        # add lines to the init method
        # title, text, link, and tags. It also calculates a meta field with date, time, and timezone
        self.title = note_dict['title']
        self.text = note_dict['text']
        self.link = note_dict['link']
        self.tags = note_dict['tags']
        self.meta = note_dict['meta']
        # turn all of this into a dictionary
        
    def as_dict(self):
        return {
            'title': self.title,
            'text': self.text,
            'link': self.link,
            'tags': self.tags,
            'meta': self.meta
        }
    
# main execution

if __name__ == '__main__':
    
    root = MainWindow() # this creates a notebook / main window called main_window. You may change the name if you want
    root.get_title()

    root.mainloop()
```
